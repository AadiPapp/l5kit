# Visualisation Example

This notebook shows some of the visulisation utility of our toolkit.

The core packages for visualisation are:
### `rasterization`
contains classes for getting visual data as multi-channels tensor and to bake them into an interpretable RGB image.
Every class has at least a `rasterize` method to get the tensor and a `to_rgb` method to convert it into and image
Few examples are:
- `BoxRasterizer`: this object renders agents (e.g. vehicles or pedestrians) as 2D oriented boxes;
- `SatelliteRasterizer`: this object renders an oriented crop from a satellite map;

### `visualization`
contains utility to draw additional information (e.g. trajectories) onto RGB images. These utilities are commonly used after a `to_rgb` call to add other information for the final visualisation. Few Examples are:
- `draw_trajectory`: this function draws 2D trajectories from coordinates and yaws offset on an image


```python
import matplotlib.pyplot as plt
import numpy as np

from avkit.data import ChunkedStateDataset, LocalDataManager
from avkit.dataset import EgoDataset, AgentDataset

from avkit.rasterization import build_rasterizer
from avkit.configs import load_config_data
from avkit.kinematic import AckermanPerturbation
from avkit.random import GaussianRandomGenerator
from avkit.visualization import draw_trajectory, TARGET_POINTS_COLOR

import os
```

### First let's configure where our data live
The data are expected to live in a folder that can be configured using the `AVKIT_DATA_FOLDER` env variable. In this exampe, the env variable is used. Data are not only raw `.zarr`, but also satellite and semantic map information.

Also, we built our code to work with a human-readable `yaml` config. This config file holds many useful information, but we will cover only few of them here

```python
# set env variable for data
os.environ["AVKIT_DATA_FOLDER"] = "/Users/pondruska/prediction-dataset"
# get config
cfg = load_config_data("./visualisation_config.yaml")
print(cfg)
```

### We can look into our current configuration for interesting fields

when loaded in python, or `yaml` is converted into a python `dict`. 

`raster_params` contains all the information related to how transform the 3D world into an image plane:
  - `raster_size`: the image plane size;
  - `pixel_size`: how many meters correspond to a pixel;
  - `ego_center`: our raster is centered around an agent, we can move the aghent in the image plane with this param;
  - `map_type`: the rasterizer to be employed. We currently support a satellite-based and a semantic-based one. We will look at the differences further down in this script.

```python
print(f'current raster_param: {cfg["raster_params"]}')
```

## Load the data

The same config file is also used to load the data. Every split in the data has it's own section, and multiple datasets can be used (as a whole or sliced). In this short example we will just take the first dataset from the `train` set. 

You may have also noticed that we're building a `LocalDataManager` object. This will resolve relative paths from the config using the `AVKIT_DATA_FOLDER_ENV_KEY` env variable we have just set.

```python
dm = LocalDataManager()
dataset_path = dm.require(cfg["train_data_loader"]["datasets"][0]["key"])
z = ChunkedStateDataset(dataset_path)
z.open()
```

### Core Objects
Along with the `rasterizer`, our tolkit contains other classes you may want to use while you build your solution. The `dataset` package as an example already implement `PyTorch` ready datasets, so you can hit the ground running and start coding immediately.

### Dataset package
We will use two classes from the `dataset` package for this example. Both of them can be iterated and return multi-channel images from the rasterizer along with future trajectories offsets and other information.
- `EgoDataset`: this dataset iterates over the AV annotations
- `AgentDataset`: this dataset iterates over other agents annotations

Both support multi-threading (through PyTorch DataLoader) OOB.

```python
rast = build_rasterizer(cfg, dm)
dataset = EgoDataset(cfg, z, rast)
```

## What if I want to visualise the AV?

let's get a sample from the dataset and use our `rasterizer` to get an rgb image we can plot

```python
data = dataset[500]

im = data["image"].transpose(1, 2, 0)
im = dataset.rasterizer.to_rgb(im)
center_in_pixels = np.asarray(cfg["raster_params"]["ego_center"]) * cfg["raster_params"]["raster_size"]
draw_trajectory(im, center_in_pixels, data["target_positions"], data["target_yaws"], TARGET_POINTS_COLOR)

plt.imshow(im[::-1])
plt.show()
```

## What if I want to change the rasterizer?

We can do it super easily by building a new rasterizer and a new dataset for it. In this example we change the value to `py_satellite` which render boxes on a aerial image

```python
cfg["raster_params"]["map_type"] = "py_satellite"
rast = build_rasterizer(cfg, dm)
dataset = EgoDataset(cfg, z, rast)
data = dataset[0]

im = data["image"].transpose(1, 2, 0)
im = dataset.rasterizer.to_rgb(im)
center_in_pixels = np.asarray(cfg["raster_params"]["ego_center"]) * cfg["raster_params"]["raster_size"]
draw_trajectory(im, center_in_pixels, data["target_positions"], data["target_yaws"], TARGET_POINTS_COLOR)

plt.imshow(im[::-1])
plt.show()
```

## What if I want to visualise an agent?

Glad you asked! We can just replace the `EgoDataset` with an `AgentDataset`. Now we're iterating over agents and not the AV anymore, and the first one happens to be the pace car (you will see this one a lot around the dataset)

```python
dataset = AgentDataset(cfg, z, rast)
data = dataset[0]

im = data["image"].transpose(1, 2, 0)
im = dataset.rasterizer.to_rgb(im)
center_in_pixels = np.asarray(cfg["raster_params"]["ego_center"]) * cfg["raster_params"]["raster_size"]
draw_trajectory(im, center_in_pixels, data["target_positions"], data["target_yaws"], TARGET_POINTS_COLOR)

plt.imshow(im[::-1])
plt.show()
```

## System Origin and Orientation

At this point you may have noticed we flip the image before plotting it on the **Y axis**.

When moving from 3D to 2D we stick to a right-hand system where the origin is in the bottom-left corner with positive x axis going right and positive y axis going up the image plane. The camera is facing down the negative z axis. 

However, both `opencv` and `pyplot` place the origin in the top-left corner with positive x going right and positive y going down in the image plane. The camera is facing down the positive z axis.\n",

The flip done on the resulting image is for visualisation purposes to accommodate the difference in the two coordinate frames.

Also all our rotations are counter-clockwise for positive value of the angle.
